### Import

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from seaborn import histplot

from sklearn.preprocessing import StandardScaler  
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
````

## Database'i importluyorum

````python
df = pd.read_csv("Dos_detection_dataset.csv")
#df = df.round(4)
df.head()
````

### Boş değer
Olası hataları önlemek amaçlı boş değer var mı kontrol ediyorum.

````python
# Checking Missing values
missing_values = df.isnull().sum()
print("found" if missing_values.any() else "not found")
````

### Dataseti ayırma
Önce labela göre bir ayrım yapıyorum.  
x: input, y: output olacak şekilde  
Sonrasında train ve test olmak üzere 2'ye ayırıyorum. Oranlama olarak %80'e %20 yaptım.

````python
#x = [df[col].tolist() for col in df.columns if col != "label"]
#x = list(zip(*x))  # or use np.array(x).T.tolist()
# Separate features and labels
x = df.drop(columns=["label"]).values
#x = df["traffic_flow"].values
y = df["label"].values

# Split into training and testing sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.9)
print("Shapes:", len(x_train), len(y_train), len(x_test), len(y_test))
````

### Değerlerin aralığını çizdiriyorum

````python

plt.figure(figsize=(20, 20))
alt_x = [df[col].tolist() for col in df.columns if col != "label"]
alt_x = list(zip(*x))
for i, column in enumerate(alt_x[:10]):
    plt.subplot(5, 2, i+1)
    histplot(alt_x[i], color='skyblue', stat='density', bins=10)
    plt.title(f"Distribution of {df.columns[i]}")
    plt.xlabel(f"{df.columns[i]}")
    plt.ylabel("Frequency")
    
plt.tight_layout()
plt.show()
````


### Normalization
uçuk değerleri engellemek için değerleri 0 ve 1 etrafına çekiyorum.

````python
scaler = StandardScaler()  
x_train = scaler.fit_transform(x_train)  
x_test = scaler.transform(x_test)# Create a plot
````

### Visualize Normalization
Normalization'ın getirdiği aralığı kontrol ediyorum.  

````python
plt.figure(figsize=(8, 4))

# Histogram visualization to see frequency distribution
plt.hist(x_train.ravel(), bins=100, color='skyblue', edgecolor='black')
plt.xlabel("Value Range")
plt.ylabel("Frequency")

# Show plot
plt.show()
````

````python
# Reducing to 2 dimensions using PCA
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train)

# Scatter plot of the 2 principal components
plt.figure(figsize=(10, 10))
plt.scatter(x_train_pca[:, 0], x_train_pca[:, 1], alpha=0.5)
plt.title("PCA of Standardized x_train")
plt.grid(True)
plt.show()
````

### Datasetin Pytorch'a entegrasyonu
Elimdeki verileri pytorch ile uyumlu hale getiriyorum.  
x'lerin aksine y'ler düz bir array olduğu için 2D array haline getiriyorum.

````python
x_train = torch.tensor(x_train, dtype=torch.float32)
x_test = torch.tensor(x_test, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)  # Reshape to (N,1)
y_test = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)  # Reshape to (N,1)
````

# Model
### Layers
5 layer; 1 input, 3 hidden, 1 output layer.
* fc1: Input layer, 11 nöron
* fc2: Hidden layer 1, 60 nöron
* fc3: Hidden layer 2, 60 nöron
* fc3: Hidden layer 3, 30 nöron
* fc4: Output layer, 1 nöron (binary olması açısından)
  
### Forward Passing
Fonksiyon olarak önce ReLU kullandım, kötü acc ve f1 değerleri alıyorum. Dataları yukarıda visualize edince fark ettim ki ölü nörona sebep olmuşum. Bu yüzden leaky ReLU kullanmaya karar verdim. Alttaki yazı zaten ReLU kullandığım zaman kendime not olarak yazdığım yazı.  

Output olarak 0-1 aralığı lazım olduğu için son fonksiyonda sigmoid kullandım. Sigmoid yerine yine ReLU kullanıp gelen veriyi normalize edince elde ettiğim sonuç çok güzel değildi.  
```Dead neuron var ise leaky ReLU'ya çevirebilirim```

### Diğer Model Denemeleri
Syntax şeklim: Input-Hidden-Output layerlar | batch | dropout şeklinde olacak.
1. 11-16-16-1 | batch yok | dropout yok. En fazla %60'larda acc gördüm. F1 score 0.4'lerdeydi
2. 11-32-16-8-1 | batch yok | dropout yok. Değerler öncekine benzer.  
Overfitting olduğunu fark ettim. Biraz araştırınca dropout koymaya karar verdim.
3. 11-16-16-8-1 | batch yok | dropout 0.5. Değerler aşırı fazla değişmedi. Overfitting'i bi nebze önledi
Daha dar ve geniş modeller denemeye karar verdim.  
Buradaki gibi: https://machinelearningmastery.com/building-a-binary-classification-model-in-pytorch/  
Geniş model garip değerler verdiği için dar modelden devam etme kararı aldım.
4. 11-60-60-30-1 | batch yok | dropout 0.2. Şuanki modelden tek farkı batch yok. Değerler yakın.
5. 11-60-60-30-1 | BatchNorm1d şeklinde | dropout 0.2. Şuanki modelim. Acc ve f1 değerleri hakkında aşağıda daha detaylı konuşacağım.

Bunlar dışında daha bir farklı değerde nöron batch dropout denedim. Ama onları not almadığım ve bazıları hariç bahsetmeye gerek olmadığı için buraya yazmadım.

````python
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(11, 60)
        self.bn1 = nn.BatchNorm1d(60)
        self.dropout1 = nn.Dropout(0.2)

        self.fc2 = nn.Linear(60, 60)
        self.bn2 = nn.BatchNorm1d(60)
        self.dropout2 = nn.Dropout(0.2)

        self.fc3 = nn.Linear(60, 30)
        self.bn3 = nn.BatchNorm1d(30)
        self.dropout3 = nn.Dropout(0.2)
        
        self.fc4= nn.Linear(30, 1)
        
        # Use LeakyReLU instead of ReLU to prevent dead neurons
        self.activation = nn.LeakyReLU(negative_slope=0.01)
        
    def forward(self, x):
        x = self.activation(self.bn1(self.fc1(x)))
        x = self.dropout1(x)

        x = self.activation(self.bn2(self.fc2(x)))
        x = self.dropout2(x)
        
        x = self.activation(self.bn3(self.fc3(x)))
        x = self.dropout3(x)

        x = self.fc4(x)
        x = torch.sigmoid(x)
        
        return x
````

````python
model = SimpleNN()
````

### Cost Function 
Cost function olarak ***Binary Classification*** yağtığım için **Binary Cross-Entropy** fonksiyonunu kullanacağım.  
Yukarıda bahsetmedim fakat son fonksiyon olarak sigmoid kullanmadığım durumlarda BCEWithLogitsLoss kullandım. Verdiği değerleri sevmedim o yüzden iptal ettim.

````python
# Binary Cross-Entropy Cost Function
criterion  = nn.BCELoss()
````

### Optimizer
Adam fonksiyonunu kullanıyorum.
SGD fonksiyonunu da denedim ama çok manuel kalacağı için istemedim.  
Her 2 fonksiyon için de learning rate(lr)'leri 0.1/0.01/0.001/0.0001'lerde neredeyse tüm modellerde test ettim. En iyi sonuçlardan biri şuanki kullandığım.

````python
optimizer = optim.Adam(model.parameters(), lr=0.01)
#optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)
````

### Model Eğitimi
Model eğitimi kısmı. Zaten anlaşılan bir yer. Bahsetmem gereken bir yer varsa num_epoch sayısıdır. Bu sayıyı çok arttırdığımda düzgün değerler alamadığım zamanlar oldu.  
Test ettiğim kadarıyla 50-250 arası en ideal aralık. 100'den daha iyi bir sayı vardır ama 100'de bırakmak istiyorum.

````python
num_epochs = 150
for epoch in range(num_epochs):
    optimizer.zero_grad()   # Clear previous gradients
    outputs = model(x_train)  # Forward pass
    
    loss = criterion(outputs, y_train)  # Compute loss
    loss.backward()  # Backpropagation (compute gradients)
    optimizer.step()  # Update weights

    if (epoch + 1) % 10 == 0:  # Print every 10 epochs
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
````

### Burası sigmoid kullanmadığım durumda son değeri normalize ettiğim yer.

### Skorlar
Son olarak skorları gördüğüm yerler buralar. Overfitting sorunum var şuanki modelde.  
En yüksek training değeri olarak %95 acc, 0.9 f1 score gördüm. Fakat o modelde validation acc %60, f1 skor 0.4'lerde.
Aldığım en yükse test f1 skoru 0.7'lerde fakat o model daha basit bir modeldi batch dropout vs. yoktu. Bu projeyi bir öğrenme projesi olarak gördüğüm için o modeli iptal ettim. Şuanki hali ondan daha kompleks fakat skor biraz daha az. Bu modelde en yüksek %65 val acc, 0.6 f1 skor gördüm.


Threshold değeri olarak 0.5'den 0.2' çektiğimde training acc ve f1 skorları %10'luk fark ederken Test skorlarım neredeyse 2 katına çıktı.  
Daha alt threshold değerinde training çöküyor. O yüzden bu şekil bırakıyorum.

````python
threshold = 0.2
with torch.no_grad():  # Disable gradient computation for evaluation
    y_pred_train = model(x_train)
    rounded_y_pred_train = (y_pred_train > threshold).float()  # Convert probabilities to 0/1 labels
    train_accuracy = (rounded_y_pred_train == y_train).float().mean()
    print(f"Training Accuracy: {train_accuracy.item() * 100:.2f}%")
    
    y_pred_train_np = rounded_y_pred_train.cpu().numpy()
    y_train_np = y_train.cpu().numpy()

    f1_train = f1_score(y_train_np, y_pred_train_np)
    print(f"Train F1-Score: {f1_train:.4f}")
    

    y_pred_test = model(x_test)
    rounded_y_pred_test = (y_pred_test > threshold).float()
    validation_accuracy = (rounded_y_pred_test == y_test).float().mean()
    print(f"Validation Accuracy: {validation_accuracy.item() * 100:.2f}%")
    
    y_pred_test_np = rounded_y_pred_test.cpu().numpy()
    y_test_np = y_test.cpu().numpy()

    f1_test = f1_score(y_test_np, y_pred_test_np)
    print(f"Test F1-Score: {f1_test:.4f}")
    
    print("Class 0 count:", (y_pred_test_np == 0).sum())
    print("Class 1 count:", (y_pred_test_np == 1).sum())
````
